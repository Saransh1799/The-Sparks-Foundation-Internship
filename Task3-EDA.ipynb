### Distribution Plots

```python
fig, axs = plt.subplots(ncols=2, nrows=2,figsize=(10,10))
sns.distplot(data['Sales'], color = 'Brown', ax=axs[0][0])
sns.distplot(data['Profit'], color = 'Green', ax=axs[0][1])
sns.distplot(data['Quantity'], color = 'Red', ax=axs[1][0])
sns.distplot(data['Discount'], color = 'Purple', ax=axs[1][1])


axs[0][0].set_title('Sales Distribution', fontsize = 20  )
axs[0][1].set_title('Profit Distribution', fontsize = 20 )
axs[1][0].set_title('Quantity Distribution', fontsize = 20 )
axs[1][1].set_title('Discount Distribution', fontsize = 20 )

plt.show()
```

### Profit vs Discount Relation

```python
plt.figure(figsize = (10, 5))
# profit/loss by Discount level
sns.lineplot('Discount', 'Profit', data = data, color = 'Green', label = 'Discount Level')
plt.ylabel('Profit')
plt.title('Profit/Loss by varying Discount ', fontsize = 18)
plt.show()
```

### The lineplot concludes that :
1.The discount level should be lowered to 10% or less in order to increase the profit.

2.The worst losses are incurred when discount approximates  around 50%. This clearly indicates end-of-season sales, and sales for older design pieces or defective products.

3.Superstore is most profitable when the discount level reaches lower than 10%. This allows less occurences of lowered prices after decreasing the discount level, and thus more profits leading to better business exposure.
